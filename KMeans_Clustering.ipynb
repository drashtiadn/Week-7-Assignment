## K-Means Clustering

#### 1. Exploratory Data Analysis:

```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset
dataset = pd.read_csv("Mall_Customers.csv")

# Display the first few rows of the dataset
print(dataset.head())
X = dataset.iloc[:, [3,4]].values
```

```python
# Check the basic statistics of the numerical features
print(dataset.describe())
```

```python
# Plot the distributions of numerical features
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(dataset['Age'], kde = True)
plt.subplot(2, 2, 2)
sns.histplot(dataset['Annual Income (k$)'], kde = True)
plt.subplot(2, 2, 3)
sns.histplot(dataset['Spending Score (1-100)'], kde = True)
plt.subplot(2, 2, 4)
sns.countplot(dataset['Gender'])
plt.show()
```

```python
# Check correlations between numerical features
correlation = dataset.corr()
sns.heatmap(correlation, annot = True)
plt.show()
```

#### 2. Data Pre-processing:

```python
# Convert 'Gender' column to numerical values (0 for Female, 1 for Male)
dataset['Gender'] = dataset['Gender'].map({'Female': 0, 'Male': 1})
print(dataset.head())
```

```python
# Checking missing values
dataset.isnull().sum()
```

#### 3. K-means Clustering:

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans  = KMeans(n_clusters = i, init = 'k-means++', random_state = 5)          
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Training the K-Means model on the dataset
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 5)
y_kmeans = kmeans.fit_predict(X)
```

```python
print(y_kmeans)
```

```python
# Visualising the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
